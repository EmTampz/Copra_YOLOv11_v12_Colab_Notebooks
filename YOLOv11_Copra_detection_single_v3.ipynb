```python
# ======================================================================
#  COPRA FINAL THESIS PIPELINE (TRAIN/VAL + TEST METRICS + F1)
# ======================================================================

import os, time, glob, json, yaml, zipfile
from datetime import datetime
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

from ultralytics import YOLO
from roboflow import Roboflow
import wandb
from IPython.display import display
from matplotlib.colors import LinearSegmentedColormap

# ---------------------------------------------------------------
# 1. CONFIGURATION
# ---------------------------------------------------------------
CONFIG = dict(
    epochs=100,          # e:
    batch=16,            # b:
    lr0=0.001,           # lr
    patience=50,          # p
    optimizer="auto",    # auto
    imgsz=640,           # img
)

# ---------------------------------------------------------------
# 2. SETUP & DRIVE MOUNT
# ---------------------------------------------------------------
try:
    from google.colab import drive
    drive.mount('/content/drive')
    DRIVE_OK = True
    print("Google Drive mounted successfully.")
except:
    DRIVE_OK = False
    print("Google Drive not mounted. Results will be local only.")

# --- BASE DIRECTORIES ---
BASE_DIR = "/content/drive/MyDrive/yolo_copra" if DRIVE_OK else "/content/yolo_copra"
RESULTS_DIR = os.path.join(BASE_DIR, "copra_yolo_final_pipeline_3")
os.makedirs(RESULTS_DIR, exist_ok=True)

# --- WANDB ---
os.environ["WANDB_API_KEY"] = "02a2986eed98555093c4c2920a04c631a89430cc"
try:
    import wandb
    wandb.login(key=os.environ["WANDB_API_KEY"], relogin=True)
    WANDB_PROJECT = "Copra-YOLO-Final"
except:
    print("W&B not installed or login failed.")

# ---------------------------------------------------------------
# LOAD DATASET
# ---------------------------------------------------------------
rf = Roboflow(api_key="LS1Ohi2OQZ178OOqQ4zc")
project = rf.workspace("cv-opfhv").project("mor-v6-jj3ic")
dataset = project.version(2).download("yolov11")
DATA_YAML = os.path.join(dataset.location, "data.yaml")

with open(DATA_YAML, "r") as f:
    meta = yaml.safe_load(f)

class_names = meta["names"]
num_classes = len(class_names)
# Colors for plots
COPRA = {"cream": "#FFE4C2", "soft_orange": "#FFB56B", "orange": "#FF914D", "red_orange": "#E85C29", "husk_brown": "#8B4513", "shell_brown": "#5A2E0A", "burnt_dark": "#3B1F0E"}

# ======================================================================
# HELPER â€” LOG IMAGES TO W&B
# ======================================================================
def wandb_log_image(path, name):
    if os.path.exists(path):
        try:
            wandb.log({name: wandb.Image(path)})
        except Exception as e:
            print(f" W&B log failed for {name}: {e}")

# ======================================================================
#  TRAIN
# ======================================================================
MODEL_PATH = "yolo11n.pt"

def train_yolo(run_dir, resume=False):
    print(f"\n STARTING TRAINING IN: {run_dir}")

    start_time = time.time()

    model_to_load = MODEL_PATH
    if resume:
        weights_dir = os.path.join(run_dir, "train", "weights")
        last_ckpt = os.path.join(weights_dir, "last.pt")
        if os.path.exists(last_ckpt):
            model_to_load = last_ckpt
            print(f"   Resuming from: {last_ckpt}")
        else:
            print("   Resume requested but no checkpoint found. Starting fresh.")
            resume = False

    model = YOLO(model_to_load)

    try:
        model.train(
            data=DATA_YAML,
            epochs=CONFIG["epochs"],
            batch=CONFIG["batch"],
            lr0=CONFIG["lr0"],
            patience=CONFIG["patience"],
            optimizer=CONFIG["optimizer"],
            imgsz=CONFIG["imgsz"],
            project=run_dir,
            name="train",
            exist_ok=True,
            resume=resume,
            pretrained=(not resume),
            val=True,
            save_period=5
        )
    except AssertionError as e:
        # If Ultralytics says "training is finished", we catch it and skip to testing.
        if "finished" in str(e) or "nothing to resume" in str(e):
            print(f"\n Resuming check: {e}")
            print(" Training is already complete.")
            print("Proceeding to Validation/Testing Phase.")
            return run_dir
        else:
            # If it's a real crash (e.g. memory error), raise it.
            raise e


    end_time = time.time()
    total_seconds = end_time - start_time
    from datetime import timedelta
    duration_str = str(timedelta(seconds=int(total_seconds)))

    print(f"    Training finished. Total Time: {duration_str}")
    print(f"    Weights at: {run_dir}/train/weights/")

    # Save Time Log
    time_file_path = os.path.join(run_dir, "training_time.txt")
    with open(time_file_path, "w") as f:
        f.write(f"Training Started: {time.ctime(start_time)}\n")
        f.write(f"Training Ended:   {time.ctime(end_time)}\n")
        f.write(f"Duration: {duration_str}\n")
        f.write(f"Epochs: {CONFIG['epochs']}\n")
        f.write(f"Batch:  {CONFIG['batch']}\n")

    print(f"Time log saved to: {time_file_path}")

    return run_dir

# ======================================================================
#  CHECK STATUS
# ======================================================================
def check_training_status():
    """
    Scans RESULTS_DIR to find the latest run.
    Returns: (run_dir, needs_training)
    """
    # 1. Create main dir if missing
    if not os.path.exists(RESULTS_DIR):
        os.makedirs(RESULTS_DIR, exist_ok=True)
        print("No previous runs found (Main folder created). New training required.")
        return None, True

    # 2. Scan for existing run folders
    run_dirs = sorted(
        os.path.join(RESULTS_DIR, d)
        for d in os.listdir(RESULTS_DIR)
        if os.path.isdir(os.path.join(RESULTS_DIR, d))
    )

    # 3. If no runs exist, start fresh
    if not run_dirs:
        print("No previous runs found. New training required.")
        return None, True

    # 4. Grab the latest run
    latest_run_dir = run_dirs[-1]
    print(f"ğŸ” Found existing run: {latest_run_dir}")

    # 5. Check if it's finished (best.pt exists)
    best_pt = os.path.join(latest_run_dir, "train", "weights", "best.pt")
    if os.path.exists(best_pt):
        print(f"Found 'best.pt'. Training finished.")
        return latest_run_dir, False # Run exists, Training NOT needed

    # 6. Check if it crashed (results.csv exists but no best.pt)
    results_file = os.path.join(latest_run_dir, "train", "results.csv")
    if os.path.exists(results_file):
        print(" 'best.pt' missing but results.csv exists. Run likely crashed. Resuming.")
        return latest_run_dir, True # Run exists, Training NEEDED (Resume)

    # 7. Empty folder (Restart in this folder)
    print("Folder exists but looks empty. Starting fresh in this folder.")
    return latest_run_dir, True

# ======================================================================
#  PLOT CONSOLIDATED METRICS
# ======================================================================
def plot_training_metrics(run_dir):
    print("\n GENERATING MASTER TRAINING DASHBOARD...")
    results_csv = os.path.join(run_dir, "train", "results.csv")

    if not os.path.exists(results_csv):
        print(f" No results.csv found at {results_csv}. Cannot plot metrics.")
        return

    try:
        df = pd.read_csv(results_csv)
        df.columns = df.columns.str.strip()

        fig = plt.figure(figsize=(20, 12))

        # Row 1: Losses
        ax1 = plt.subplot(2, 3, 1)
        ax1.plot(df['train/box_loss'], label='Train', color=COPRA['orange'], linewidth=2)
        ax1.plot(df['val/box_loss'], label='Val', color=COPRA['burnt_dark'], linewidth=2, linestyle='--')
        ax1.set_title("Box Loss", weight='bold', color=COPRA['shell_brown']); ax1.legend(); ax1.grid(alpha=0.3)

        ax2 = plt.subplot(2, 3, 2)
        ax2.plot(df['train/cls_loss'], label='Train', color=COPRA['orange'], linewidth=2)
        ax2.plot(df['val/cls_loss'], label='Val', color=COPRA['burnt_dark'], linewidth=2, linestyle='--')
        ax2.set_title("Class Loss", weight='bold', color=COPRA['shell_brown']); ax2.legend(); ax2.grid(alpha=0.3)

        ax3 = plt.subplot(2, 3, 3)
        ax3.plot(df['train/dfl_loss'], label='Train', color=COPRA['orange'], linewidth=2)
        ax3.plot(df['val/dfl_loss'], label='Val', color=COPRA['burnt_dark'], linewidth=2, linestyle='--')
        ax3.set_title("DFL Loss", weight='bold', color=COPRA['shell_brown']); ax3.legend(); ax3.grid(alpha=0.3)

        # Row 2: Metrics
        ax4 = plt.subplot(2, 2, 3)
        ax4.plot(df['metrics/precision(B)'], color=COPRA['red_orange'], linewidth=2, label='Precision')
        ax4.plot(df['metrics/recall(B)'], color=COPRA['soft_orange'], linewidth=2, label='Recall')
        ax4.set_title("Precision & Recall", weight='bold', color=COPRA['shell_brown']); ax4.set_ylim(0, 1.05); ax4.legend(); ax4.grid(alpha=0.3)

        ax5 = plt.subplot(2, 2, 4)
        ax5.plot(df['metrics/mAP50(B)'], color=COPRA['orange'], linewidth=2, label='mAP @ 50')
        ax5.plot(df['metrics/mAP50-95(B)'], color=COPRA['burnt_dark'], linewidth=2, label='mAP @ 50-95')
        ax5.set_title("mAP Metrics", weight='bold', color=COPRA['shell_brown']); ax5.set_ylim(0, 1.05); ax5.legend(); ax5.grid(alpha=0.3)

        plt.suptitle(f"Consolidated Training Metrics", fontsize=18, color=COPRA['shell_brown'], y=1.02)
        plt.tight_layout()

        save_path = os.path.join(run_dir, "training_metrics_consolidated.png")
        plt.savefig(save_path, bbox_inches='tight', dpi=150)
        plt.show()
        wandb_log_image(save_path, "0. Consolidated Training Metrics")
        print(f" Master metrics dashboard saved to: {save_path}")

    except Exception as e:
        print(f"Failed to plot metrics: {e}")

# ======================================================================
#  TEST & VISUALIZE (UPDATED: BAR GRAPH LOGIC + INLINE DISPLAY)
# ======================================================================
def run_test_and_visualize(run_dir):
    print("\n RUNNING TEST & VISUALIZATION...")
    best_pt = os.path.join(run_dir, "train", "weights", "best.pt")
    if not os.path.exists(best_pt):
        best_pt = os.path.join(run_dir, "train", "weights", "last.pt")
        if not os.path.exists(best_pt):
            raise FileNotFoundError(" No weights found.")

    print(f" Using model: {best_pt}")
    model = YOLO(best_pt)
    test_dir = os.path.join(run_dir, "test_results_final")
    os.makedirs(test_dir, exist_ok=True)

    # Run 1: Standard (IoU 0.5)
    metrics_50 = model.val(data=DATA_YAML, split="test", project=test_dir, name="test_run_iou50", exist_ok=True, iou=0.5, save=True, save_conf=True, plots=True)

    # Run 2: Strict (IoU 0.95)
    metrics_95 = model.val(data=DATA_YAML, split="test", project=test_dir, name="test_run_iou95", exist_ok=True, iou=0.95, save=True, save_conf=True, plots=True)

    # Extract Confusion Matrices
    cm_full_50 = metrics_50.confusion_matrix.matrix
    cm_full_95 = metrics_95.confusion_matrix.matrix

    # --- HELPER: CALCULATE METRICS DIRECTLY FROM CM (SYNC WITH BACKGROUND) ---
    # This is the critical logic to ensure Bar Graph matches Matrix
    def get_cm_metrics(cm, num_classes):
        # cm shape is (N+1, N+1) where N+1 is Background
        # TP = Diagonal
        tp = np.diag(cm)

        # FP = Sum of Column (Predicted Class) - TP
        # This includes the Background ROW (Actual Background predicted as Class X)
        fp = cm.sum(axis=0) - tp

        # FN = Sum of Row (Actual Class) - TP
        # This includes the Background COL (Actual Class X predicted as Background)
        fn = cm.sum(axis=1) - tp

        # Calculate (Add epsilon to avoid div by zero)
        eps = 1e-7
        precision = tp / (tp + fp + eps)
        recall    = tp / (tp + fn + eps)
        f1        = 2 * (precision * recall) / (precision + recall + eps)

        # Return only the Object Classes (slice off the background index at the end)
        return precision[:num_classes], recall[:num_classes], f1[:num_classes]

    # --- DERIVE METRICS FROM MATRIX ---
    # This ensures the bars exactly match the CM visual, including background errors
    p50_cm, r50_cm, f1_50_cm = get_cm_metrics(cm_full_50, num_classes)
    p95_cm, r95_cm, f1_95_cm = get_cm_metrics(cm_full_95, num_classes)

    # Standard map (still useful to keep from YOLO native)
    map5095 = metrics_50.box.maps

    # --- VISUALIZATION SETUP ---
    cm_50_no_bg = cm_full_50[:num_classes, :num_classes]
    cm_95_no_bg = cm_full_95[:num_classes, :num_classes]

    row_sums_50_true = cm_full_50.sum(axis=1)
    row_sums_95_true = cm_full_95.sum(axis=1)

    copra_colors = ["#FFE4C2", "#FFB56B", "#FF914D", "#E85C29"]
    copra_cmap = LinearSegmentedColormap.from_list("copra_oranges", copra_colors)

    def plot_cm(ax, cm, title, labels, true_sums=None):
        num_labels = len(labels)
        im = ax.imshow(cm, cmap=copra_cmap)
        ax.set_xticks(np.arange(num_labels)); ax.set_yticks(np.arange(num_labels))
        ax.set_xticklabels(labels, rotation=45, ha="right", color=COPRA["shell_brown"])
        ax.set_yticklabels(labels, color=COPRA["shell_brown"])
        ax.set_title(title, color=COPRA["husk_brown"], pad=15)

        if true_sums is not None:
            denoms = true_sums[:num_labels]
        else:
            denoms = cm.sum(axis=1)

        denoms_safe = np.where(denoms == 0, 1, denoms)
        thresh = (cm.max() + cm.min()) / 2.0

        for i in range(num_labels):
            for j in range(num_labels):
                val = int(cm[i, j])
                percent = val / denoms_safe[i] if denoms_safe[i] > 0 else 0
                text_label = f"{val}\n({percent:.1%})"
                text_color = COPRA["burnt_dark"] if val < thresh else COPRA["cream"]
                ax.text(j, i, text_label, ha="center", va="center", color=text_color, fontsize=9, weight="bold")

    # --- PICTURE 1: NO BACKGROUND (IoU 0.5 and IoU 0.95) ---
    fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    plot_cm(ax1, cm_50_no_bg, "Confusion Matrix (IoU 0.5) ", class_names, true_sums=row_sums_50_true)
    plot_cm(ax2, cm_95_no_bg, "Confusion Matrix (IoU 0.95) ", class_names, true_sums=row_sums_95_true)
    plt.suptitle("Detection Performance (No Background)", fontsize=18, color=COPRA["shell_brown"], y=1.05)
    plt.tight_layout()
    save_path_no_bg = os.path.join(test_dir, "cm_comparison_NO_bg.png")
    plt.savefig(save_path_no_bg)
    plt.show() # <--- INLINE DISPLAY ADDED
    plt.close()
    wandb_log_image(save_path_no_bg, "2a. Confusion Matrix No Background")
    print(f"Saved Confusion Matrix (No BG) to: {save_path_no_bg}")

    # --- PICTURE 2: WITH BACKGROUND (IoU 0.5 and IoU 0.95) ---
    fig2, (ax3, ax4) = plt.subplots(1, 2, figsize=(16, 7))
    labels_with_bg = class_names + ["background"]
    plot_cm(ax3, cm_full_50, "Confusion Matrix (IoU 0.5) ", labels_with_bg, true_sums=row_sums_50_true)
    plot_cm(ax4, cm_full_95, "Confusion Matrix (IoU 0.95) ", labels_with_bg, true_sums=row_sums_95_true)
    plt.suptitle("Detection Performance (With Background)", fontsize=18, color=COPRA["shell_brown"], y=1.05)
    plt.tight_layout()
    save_path_with_bg = os.path.join(test_dir, "cm_comparison_WITH_bg.png")
    plt.savefig(save_path_with_bg)
    plt.show() # <--- INLINE DISPLAY ADDED
    plt.close()
    wandb_log_image(save_path_with_bg, "2b. CM With Background")
    print(f"Saved CM (With BG) to: {save_path_with_bg}")

    # --- PICTURE 3: BAR CHARTS (3-WAY COMPARISON - NOW SYNCED) ---
    def plot_grouped_bar(ax, metrics_dict, title):
        x = np.arange(len(class_names)); width = 0.25; multiplier = 0
        for name, (data, color) in metrics_dict.items():
            offset = width * multiplier
            rects = ax.bar(x + offset, data, width, label=name, color=color)
            ax.bar_label(rects, padding=3, fmt='%.2f', fontsize=8)
            multiplier += 1
        ax.set_ylabel('Score', color=COPRA["husk_brown"])
        ax.set_title(title, color=COPRA["husk_brown"], pad=15, fontsize=14)
        ax.set_xticks(x + width, class_names, color=COPRA["orange"])
        ax.legend(loc='upper left', ncols=3)
        ax.set_ylim(0, 1.1)
        ax.grid(axis='y', alpha=0.3)

    fig_bars, (ax_bar1, ax_bar2, ax_bar3) = plt.subplots(1, 3, figsize=(26, 6))

    # 1. IoU 0.5 Metrics (Using CM Derived values)
    data_05 = {
        'Precision': (p50_cm, COPRA["soft_orange"]),
        'Recall': (r50_cm, COPRA["orange"]),
        'F1': (f1_50_cm, COPRA["red_orange"])
    }
    plot_grouped_bar(ax_bar1, data_05, f"Metrics (IoU 0.5) ")

    # 2. IoU 0.5-0.95 Metrics (mAP remains native as it's an integral)
    data_avg = {
        'mAP 50-95': (map5095, COPRA["shell_brown"])
    }
    plot_grouped_bar(ax_bar2, data_avg, f"Metrics (IoU 0.5-0.95)")

    # 3. IoU 0.95 Metrics (Using CM Derived values - ESPECIALLY THIS ONE)
    data_95 = {
        'Precision': (p95_cm, COPRA["soft_orange"]),
        'Recall': (r95_cm, COPRA["orange"]),
        'F1': (f1_95_cm, COPRA["red_orange"])
    }
    plot_grouped_bar(ax_bar3, data_95, f"Metrics (IoU 0.95) ")

    plt.suptitle(f"Per-Class Metrics)", fontsize=18, color=COPRA["shell_brown"], y=1.05)
    plt.tight_layout()

    save_path_bars = os.path.join(test_dir, "per_class_metrics_bars_3way.png")
    plt.savefig(save_path_bars)
    plt.show() # <--- INLINE DISPLAY ADDED
    plt.close()
    wandb_log_image(save_path_bars, "3. Bar Metrics 3-Way")
    print(f"Saved 3-Way Bar Charts to: {save_path_bars}")

    # --- CSV ---
    df = pd.DataFrame({
        "Class": class_names,
        "Precision_0.5": p50_cm, "Recall_0.5": r50_cm, "F1_0.5": f1_50_cm,
        "mAP_50-95": map5095,
        "Precision_0.95": p95_cm, "Recall_0.95": r95_cm, "F1_0.95": f1_95_cm
    })
    df.to_csv(os.path.join(test_dir, "test_metrics_native.csv"), index=False)
    display(df)

    # --- Extract Final Test mAP Values ---
    final_map50 = metrics_50.box.map50  # Overall mAP@50
    final_map5095 = metrics_50.box.map  # Overall mAP@50-95
    final_precision = metrics_50.box.mp # Overall Precision
    final_recall = metrics_50.box.mr    # Overall Recall
    final_f1 = metrics_50.box.f1.mean() # Overall F1 Score (Scalar)

    return test_dir, final_map50, final_map5095, final_precision, final_recall, final_f1

# ======================================================================
#  SAVE ALL PREDICTIONS
# ======================================================================
def save_all_test_predictions(run_dir, test_dir):
    print("\n SAVING RAW INFERENCE (NMS DISABLED / RELAXED)...")
    best_pt = os.path.join(run_dir, "train", "weights", "best.pt")
    if not os.path.exists(best_pt):
        best_pt = os.path.join(run_dir, "train", "weights", "last.pt")

    model = YOLO(best_pt)
    data_yaml_dir = os.path.dirname(DATA_YAML)
    test_images_source = os.path.join(data_yaml_dir, "test", "images")

    if not os.path.isdir(test_images_source):
        print(f" Error: Test folder not found at {test_images_source}")
        return

    # Folder specifically for this experiment
    pred_save_dir = os.path.join(test_dir, "experiment_nms_off")

    model.predict(
        source=test_images_source,
        project=pred_save_dir,
        name="predict_run",
        exist_ok=True,
        save=True,
        save_txt=True,
        save_conf=True,

        # --- THE EXPERIMENT SETTINGS ---
        conf=0.1,           # LOWER this to see even low-confidence "ghosts"
        iou=0.95,           # HIGHER this (0.9 - 1.0) to keep overlapping boxes
        agnostic_nms=False, # Ensure different classes (Whole vs Part) don't delete each other
        max_det=300         # Allow many detections per image
    )
    print(f" Experimental inference saved to: {os.path.join(pred_save_dir, 'predict_run')}")

# ======================================================================
# MAIN EXECUTION
# ======================================================================

# 1. INTELLIGENT CHECK: Scan for existing runs FIRST
run_dir, needs_training = check_training_status()

# 2. RUN NAME SETUP
if run_dir:
    # If we found an existing run, USE IT (Preserve the old timestamp)
    run_name = os.path.basename(run_dir)
    print(f"RESUMING/USING EXISTING FOLDER: {run_name}")
else:
    # Only generate a NEW timestamp if no runs were found
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    run_name = (f"e{CONFIG['epochs']}_p{CONFIG['patience']}_b{CONFIG['batch']}_"
                f"lr{CONFIG['lr0']}_{CONFIG['optimizer']}_img{CONFIG['imgsz']}_{timestamp}")
    run_dir = os.path.join(RESULTS_DIR, run_name)
    print(f"CREATING NEW RUN: {run_name}")

# 3. Init WandB
wandb.init(project=WANDB_PROJECT, name=run_name, resume="allow", config=CONFIG)

# 4. Train
if needs_training:
    # Resume=True if the folder actually exists
    run_dir = train_yolo(run_dir, resume=os.path.exists(run_dir))
else:
    print("\n Skipping training phase (Model already finished).")
# 5. Train
if needs_training:
    run_dir = train_yolo(run_dir, resume=True if os.path.exists(os.path.join(run_dir, "train")) else False)
else:
    print("\n Skipping training phase (Model already finished).")

# 6. Plot Metrics
plot_training_metrics(run_dir)

# 7. Test & Final Data Summary
try:
    # Unpack ALL 5 values now (Added F1)
    test_dir, test_map50, test_map5095, test_prec, test_rec, test_f1 = run_test_and_visualize(run_dir)

    # Run the Experimental Prediction Save
    save_all_test_predictions(run_dir, test_dir)

    # --- FINAL DATA EXTRACTION & SAVE TO DRIVE ---
    results_path = os.path.join(run_dir, "train", "results.csv")
    time_path = os.path.join(run_dir, "training_time.txt")
    summary_save_path = os.path.join(run_dir, "thesis_summary_metrics.txt")

    print("\n" + "="*50)
    print(f"THESIS DATA SUMMARY FOR: {run_name}")
    print("="*50)

    summary_text = []
    summary_text.append(f"Run Name: {run_name}")
    summary_text.append("="*50)

    # SECTION 1: TEST SET RESULTS (From best.pt)
    summary_text.append("[TEST RESULTS ]")
    line1 = f"mAP @ 50-95 : {test_map5095:.5f}"
    line2 = f"mAP @ 50    : {test_map50:.5f}"
    line3 = f"Precision   : {test_prec:.5f}"
    line4 = f"Recall      : {test_rec:.5f}"
    line5 = f"F1 Score    : {test_f1:.5f}"

    print(line1); summary_text.append(line1)
    print(line2); summary_text.append(line2)
    print(line3); summary_text.append(line3)
    print(line4); summary_text.append(line4)
    print(line5); summary_text.append(line5)
    summary_text.append("-" * 20)

    # SECTION 2: TRAINING/VAL RESULTS (From results.csv)
    if os.path.exists(results_path):
        res = pd.read_csv(results_path)
        res.columns = res.columns.str.strip()
        last_row = res.iloc[-1]

        summary_text.append("[TRAIN/VAL RESULTS]")
        val_p = f"/n Train/Val Precision: {last_row['metrics/precision(B)']:.5f}"
        val_r = f"Train/Val Recall:    {last_row['metrics/recall(B)']:.5f}"
        val_m = f"Train/Val mAP50:     {last_row['metrics/mAP50(B)']:.5f}"
        val_m95 = f"Train/Val mAP50-95:  {last_row['metrics/mAP50-95(B)']:.5f}"

        print(val_p); summary_text.append(val_p)
        print(val_r); summary_text.append(val_r)
        print(val_m); summary_text.append(val_m)
        print(val_m95); summary_text.append(val_m95)

    if os.path.exists(time_path):
        with open(time_path, 'r') as f:
            content = f.read()
            t_val = content.split('Duration:')[1].strip() if "Duration:" in content else content.strip()
            line_time = f"\nâ± Training Time:        {t_val}"
            print(line_time); summary_text.append(line_time)

    # SAVE TO DRIVE
    with open(summary_save_path, "w") as f:
        f.write("\n".join(summary_text))

    print("="*50)
    print(f" Summary Block saved to: {summary_save_path}")

except Exception as e:
    print(f"\n Visualization failed: {e}")

wandb.finish()
```
