```python

import os, time, glob, json, yaml, zipfile
from datetime import datetime
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

from ultralytics import YOLO
from roboflow import Roboflow
import wandb
from IPython.display import display

# ---------------------------------------------------------------
# COLORS ‚Äî COPRA+ PALETTE
# ---------------------------------------------------------------
COPRA = {
    "cream": "#FFE4C2",
    "soft_orange": "#FFB56B",
    "orange": "#FF914D",
    "red_orange": "#E85C29",
    "husk_brown": "#8B4513",
    "shell_brown": "#5A2E0A",
    "burnt_dark": "#3B1F0E"
}


# ---------------------------------------------------------------
# 1. SETUP
# ---------------------------------------------------------------
try:
    from google.colab import drive
    drive.mount('/content/drive')
    DRIVE_OK = True
except:
    DRIVE_OK = False

BASE_DIR = "/content/drive/MyDrive/yolo_copra" if DRIVE_OK else "/content/yolo_copra"
RESULTS_DIR = os.path.join(BASE_DIR, "copra_yolo_final_pipeline")
os.makedirs(RESULTS_DIR, exist_ok=True)

# --- WANDB (SECURITY WARNING) ---
os.environ["WANDB_API_KEY"] = "02a2986eed98555093c4c2920a04c631a89430cc"
!wandb login $WANDB_API_KEY --relogin
WANDB_PROJECT = "Copra-YOLO-Final"

# ---------------------------------------------------------------
# LOAD DATASET
# ---------------------------------------------------------------
rf = Roboflow(api_key="LS1Ohi2OQZ178OOqQ4zc")
project = rf.workspace("cv-opfhv").project("mor-v6-jj3ic")
dataset = project.version(2).download("yolov11")
DATA_YAML = os.path.join(dataset.location, "data.yaml")


with open(DATA_YAML, "r") as f:
    meta = yaml.safe_load(f)

class_names = meta["names"]
num_classes = len(class_names)

# ======================================================================
# HELPER ‚Äî LOG IMAGES TO W&B
# ======================================================================
def wandb_log_image(path, name):
    """Logs an image file to W&B."""
    if os.path.exists(path):
        try:
            wandb.log({name: wandb.Image(path)})
        except Exception as e:
            print(f"‚ö†Ô∏è W&B log failed for {name}: {e}")
    else:
        print(f"‚ö†Ô∏è W&B log skipped (file not found): {path}")

# ======================================================================
#  TRAIN
# ======================================================================
CONFIG = dict(
    epochs=200,
    batch=16,
    lr0=0.001,
    patience=50,
    optimizer="auto",
    imgsz=640,
)

MODEL_PATH = "yolo11n.pt"

def train_yolo():
    existing = sorted([d for d in os.listdir(RESULTS_DIR)
                       if os.path.isdir(os.path.join(RESULTS_DIR, d))])

    resume, run_name = False, None
    model_to_load = MODEL_PATH

    if existing:
        last_dir = os.path.join(RESULTS_DIR, existing[-1])
        ckpt = latest_checkpoint(last_dir)
        if ckpt:
            resume = True
            run_name = os.path.basename(last_dir)
            model_to_load = ckpt

    if not run_name:
        run_name = (
            f"e{CONFIG['epochs']}_p{CONFIG['patience']}"
            f"_b{CONFIG['batch']}_lr{CONFIG['lr0']}_"
            f"{CONFIG['optimizer']}_img{CONFIG['imgsz']}_"
            f"{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        )

    run_dir = os.path.join(RESULTS_DIR, run_name)
    os.makedirs(run_dir, exist_ok=True)

    wandb.init(project=WANDB_PROJECT, name=run_name, resume="allow", config=CONFIG)

    model = YOLO(model_to_load)

    model.train(
        data=DATA_YAML,
        epochs=CONFIG["epochs"],
        batch=CONFIG["batch"],
        lr0=CONFIG["lr0"],
        patience=CONFIG["patience"],
        optimizer=CONFIG["optimizer"],
        imgsz=CONFIG["imgsz"],
        project=run_dir,
        name="train",
        exist_ok=True,
        resume=resume,
        pretrained=bool(resume),
        val=True,
        save_period=5
    )

    print(f"üèÅ Training complete. Weights at: {run_dir}/train/weights/")
    return run_dir
# ======================================================================
#  AUTO RESUME (CORRECTED)
# ======================================================================
def latest_checkpoint(run_dir):
    weights_dir = os.path.join(run_dir, "train", "weights")
    ckpts = glob.glob(os.path.join(weights_dir, "*.pt"))
    return max(ckpts, key=os.path.getmtime) if ckpts else None

def auto_resume():
    """
    Checks the latest run's results.csv to see if the
    configured number of epochs has been completed.
    """
    run_dirs = sorted(
        os.path.join(RESULTS_DIR, d)
        for d in os.listdir(RESULTS_DIR)
        if os.path.isdir(os.path.join(RESULTS_DIR, d))
    )

    if not run_dirs:
        # Case 1: No runs found at all
        print("No runs found ‚Üí Training new model...")
        return train_yolo()

    # --- [NEW LOGIC] ---
    latest_run_dir = run_dirs[-1]
    results_file = os.path.join(latest_run_dir, "train", "results.csv")

    # Get the target number of epochs from the global config
    target_epochs = CONFIG['epochs']

    if os.path.exists(results_file):
        try:
            # Read the CSV file to check progress
            df = pd.read_csv(results_file)

            # Clean up column names (sometimes they have whitespace)
            df.columns = df.columns.str.strip()

            if not df.empty:
                # Get the last epoch number (it's 1-based)
                last_epoch_trained = df['epoch'].iloc[-1]

                if last_epoch_trained >= target_epochs:
                    # Case 2: Training is complete
                    print(f"‚úÖ Training complete. (Found {last_epoch_trained}/{target_epochs} epochs in {results_file})")
                    print(f"   Proceeding to visualization for: {latest_run_dir}")
                    return latest_run_dir
                else:
                    # Case 3: Training is incomplete
                    print(f"‚ö†Ô∏è Training incomplete. (Found {last_epoch_trained}/{target_epochs} epochs).")
                    print(f"   Resuming training for: {latest_run_dir}")
                    return train_yolo()
            else:
                # Case 4: results.csv is empty
                print(f"‚ö†Ô∏è Found empty results.csv. Resuming/starting training.")
                return train_yolo()

        except Exception as e:
            # Case 5: CSV is corrupted or unreadable
            print(f"‚ö†Ô∏è Error reading {results_file}: {e}. Assuming incomplete.")
            print("   Resuming training.")
            return train_yolo()
    else:
        # Case 6: No results.csv, but a checkpoint might exist
        print(f"No results.csv found in {latest_run_dir}.")
        print("   Resuming training.")
        return train_yolo()

# ======================================================================
#  TEST & VISUALIZE (MODIFIED FUNCTION)
# ======================================================================
def run_test_and_visualize(run_dir):
    print("\nüß™ RUNNING TEST & VISUALIZATION...")

    # --- Import for the custom colormap ---
    from matplotlib.colors import LinearSegmentedColormap

    best_pt = os.path.join(run_dir, "train", "weights", "best.pt")
    if not os.path.exists(best_pt):
        print("‚ö†Ô∏è best.pt not found, using last.pt")
        best_pt = os.path.join(run_dir, "train", "weights", "last.pt")
        if not os.path.exists(best_pt):
            raise FileNotFoundError("‚ùå No weights found.")

    print(f"üìå Using model: {best_pt}")
    model = YOLO(best_pt)

    test_dir = os.path.join(run_dir, "test_results_final")
    os.makedirs(test_dir, exist_ok=True)

    # --- Run 1: Get Best Case (IoU 0.5) ---
    print("\nüî¨ Running 'Best Case' Validation (Default IoU 0.5)...")
    metrics_50 = model.val(
        data=DATA_YAML,
        split="test",
        project=test_dir,
        name="test_run_iou50",
        exist_ok=True,
        iou=0.5
    )

    # --- Run 2: Get Worst Case (IoU 0.95) ---
    print("\nüî¨ Running 'Worst Case' Validation (Strict IoU 0.95)...")
    metrics_95 = model.val(
        data=DATA_YAML,
        split="test",
        project=test_dir,
        name="test_run_iou95",
        exist_ok=True,
        iou=0.95
    )

    print("‚úÖ Both validations complete. Now generating Copra+ plots...")

    # --- 1. GET ALL mAP METRICS (from the default 0.5 run) ---
    map50 = metrics_50.box.map50
    map50_95 = metrics_50.box.map
    print(f"Overall mAP@0.5: {map50:.4f}")
    print(f"Overall mAP@0.5:0.95: {map50_95:.4f}")

    # --- Get per-class metrics from BOTH runs ---
    precision_50 = metrics_50.box.p[0]
    recall_50 = metrics_50.box.r[0]
    f1_50 = metrics_50.box.f1[0]

    precision_95 = metrics_95.box.p[0]
    recall_95 = metrics_95.box.r[0]
    f1_95 = metrics_95.box.f1[0]

    # --- Get ALL confusion matrices ---
    cm_full_50 = metrics_50.confusion_matrix.matrix
    cm_50_no_bg = cm_full_50[:num_classes, :num_classes]

    cm_full_95 = metrics_95.confusion_matrix.matrix
    cm_95_no_bg = cm_full_95[:num_classes, :num_classes]

    # --------------------------------------------
    # mAP 50-95 VISUALIZATION
    # --------------------------------------------
    overall_metrics = {
        "mAP@.5": float(map50),
        "mAP@.5:.95": float(map50_95)
    }
    with open(os.path.join(test_dir, "overall_metrics.json"), "w") as f:
        json.dump(overall_metrics, f, indent=4)

    # [W&B LOG]
    wandb.log(overall_metrics)

    fig_overall, ax_overall = plt.subplots(figsize=(6, 2.5))
    ax_overall.axis('off')
    ax_overall.text(0.5, 0.7, f"mAP@0.5: {map50:.4f}", ha='center', va='center', fontsize=16, color=COPRA["husk_brown"], weight='bold')
    ax_overall.text(0.5, 0.3, f"mAP@0.5:0.95: {map50_95:.4f}", ha='center', va='center', fontsize=16, color=COPRA["husk_brown"], weight='bold')
    plt.title("Overall Model mAP Scores", color=COPRA["shell_brown"], fontsize=18)

    save_path = os.path.join(test_dir, "overall_map_metrics.png")
    plt.savefig(save_path, bbox_inches='tight', dpi=150)
    plt.show()

    # [W&B LOG]
    wandb_log_image(save_path, "1. Overall mAP Scores")


    # --------------------------------------------
    # 1x3 SIDE-BY-SIDE CONFUSION MATRICES
    # --------------------------------------------

    copra_colors = [
        "#FFE4C2",  # light
        "#FFB56B",  # soft orange
        "#FF914D",  # orange
        "#E85C29"   # red-orange (max)
    ]
    copra_cmap = LinearSegmentedColormap.from_list("copra_oranges", copra_colors)

    def plot_cm(ax, cm, title, labels):
        num_labels = len(labels)
        im = ax.imshow(cm, cmap=copra_cmap)

        ax.set_xticks(np.arange(num_labels))
        ax.set_yticks(np.arange(num_labels))
        ax.set_xticklabels(labels, rotation=45, ha="right", color=COPRA["shell_brown"])
        ax.set_yticklabels(labels, color=COPRA["shell_brown"])

        ax.set_title(title, color=COPRA["husk_brown"], pad=15)
        ax.set_xlabel("Predicted Label", color=COPRA["husk_brown"])
        ax.set_ylabel("True Label", color=COPRA["husk_brown"])

        row_sums = cm.sum(axis=1)
        row_sums_safe = np.where(row_sums == 0, 1, row_sums)

        vmax = cm.max()
        vmin = cm.min()
        if vmax == vmin:
            vmax = vmin + 1

        thresh = (vmax + vmin) / 2.0

        for i in range(num_labels):
            for j in range(num_labels):
                val = cm[i, j]
                percent = val / row_sums_safe[i] if row_sums_safe[i] > 0 else 0

                # Don't show percentage for the "background" column (j == num_classes)
                # as it's not a classification accuracy percentage
                if j == num_classes:
                    text_label = f"{val}"
                else:
                    text_label = f"{val}\n({percent:.1%})"

                text_color = COPRA["burnt_dark"] if val < thresh else COPRA["cream"]
                ax.text(j, i, text_label, ha="center", va="center",
                        color=text_color, fontsize=9, weight="bold")

    fig_cm, (ax_cm1, ax_cm2, ax_cm3) = plt.subplots(1, 3, figsize=(24, 7)) # Wider figure

    # Plot 1: Best Case (IoU 0.5, No BG)
    plot_cm(ax_cm1, cm_50_no_bg, "Confusion Matrix (Best Case: IoU 0.5)", class_names)

    # Plot 2: Worst Case (IoU 0.95, No BG)
    plot_cm(ax_cm2, cm_95_no_bg, "Confusion Matrix (Worst Case: IoU 0.95)", class_names)

    # --- [THIS IS THE CHANGE] ---
    # Plot 3: Full View (IoU 0.95, With BG)
    full_labels = class_names + ["background"]
    plot_cm(ax_cm3, cm_full_95, "Full Matrix (IoU 0.95 + Background)", full_labels)


    plt.suptitle("Best vs. Worst Case Detection Performance", fontsize=18, color=COPRA["shell_brown"], y=1.05)
    plt.tight_layout()

    save_path = os.path.join(test_dir, "confusion_matrix_comparison_full.png")
    plt.savefig(save_path)
    plt.show()

    # [W&B LOG]
    wandb_log_image(save_path, "2. Confusion Matrices (Best, Worst, Full)")


    # --------------------------------------------
    # METRICS CSV (Comparing IoU 0.5 and 0.95)
    # --------------------------------------------
    df = pd.DataFrame({
        "Class": class_names,
        "Precision_0.5": precision_50,
        "Recall_0.5": recall_50,
        "F1_0.5": f1_50,
        "Precision_0.95": precision_95,
        "Recall_0.95": recall_95,
        "F1_0.95": f1_95,
    })
    df.to_csv(os.path.join(test_dir, "test_metrics_comparison.csv"), index=False)
    print("\nPer-Class Metrics (Best vs. Worst):")
    display(df)

    # [W&B LOG]
    wandb.log({"4. Per-Class Metrics Table": wandb.Table(dataframe=df)})


    # --------------------------------------------
    # SIDE-BY-SIDE GROUPED BAR CHARTS
    # --------------------------------------------

    def plot_grouped_bar(ax, p_data, r_data, f1_data, title):
        x = np.arange(len(class_names))
        width = 0.25
        multiplier = 0

        metrics_to_plot = {
            'Precision': (p_data, COPRA["soft_orange"]),
            'Recall': (r_data, COPRA["orange"]),
            'F1-Score': (f1_data, COPRA["red_orange"]),
        }

        for metric_name, (data, color) in metrics_to_plot.items():
            offset = width * multiplier
            rects = ax.bar(x + offset, data, width, label=metric_name, color=color)
            ax.bar_label(rects, padding=3, fmt='%.2f', fontsize=8)
            multiplier += 1

        ax.set_ylabel('Score', color=COPRA["husk_brown"])
        ax.set_title(title, color=COPRA["husk_brown"], pad=15, fontsize=14)
        ax.set_xticks(x + width, class_names, color=COPRA["orange"])
        ax.legend(loc='upper left', ncols=3)
        ax.set_ylim(0, 1.1)
        ax.grid(axis='y', alpha=0.3)

    fig_bars, (ax_bar1, ax_bar2) = plt.subplots(1, 2, figsize=(20, 6))

    plot_grouped_bar(ax_bar1, precision_50, recall_50, f1_50, "Per-Class Metrics (Best Case: IoU 0.5)")
    plot_grouped_bar(ax_bar2, precision_95, recall_95, f1_95, "Per-Class Metrics (Worst Case: IoU 0.95)")

    plt.suptitle("Best vs. Worst Case Per-Class Performance", fontsize=18, color=COPRA["shell_brown"], y=1.05)
    plt.tight_layout()

    save_path = os.path.join(test_dir, "per_class_metrics_comparison.png")
    plt.savefig(save_path)
    plt.show()

    # [W&B LOG]
    wandb_log_image(save_path, "3. Per-Class Metrics (Best vs. Worst)")

    print("‚úî All Copra+ visualizations generated and logged to W&B.")
    return test_dir

# ======================================================================
#  SAVE ALL PREDICTIONS (THIS IS THE MISSING FUNCTION)
# ======================================================================
def save_all_test_predictions(run_dir, test_dir):
    """
    Runs model.predict() on the entire test set and saves all
    annotated images to a new directory.
    """
    print("\nüì∏ SAVING ALL TEST PREDICTIONS...")

    best_pt = os.path.join(run_dir, "train", "weights", "best.pt")
    if not os.path.exists(best_pt):
        print("‚ö†Ô∏è best.pt not found, using last.pt")
        best_pt = os.path.join(run_dir, "train", "weights", "last.pt")
        if not os.path.exists(best_pt):
            raise FileNotFoundError("‚ùå No weights found for prediction.")

    print(f"üìå Using model for prediction: {best_pt}")
    model = YOLO(best_pt)

    # --- [FIX] ---
    # Manually build the correct 'test/images' path from your DATA_YAML
    # to prevent errors from bad paths inside the YAML file.

    # 1. Get the root directory of your dataset (where data.yaml is)
    data_yaml_dir = os.path.dirname(DATA_YAML)

    # 2. Assume the test images are in the standard 'test/images' subfolder
    test_images_source = os.path.join(data_yaml_dir, "test", "images")

    # 3. Check if this correct path actually exists
    if not os.path.isdir(test_images_source):
        print(f"‚ùå Error: The standard test folder was not found at the expected location.")
        print(f"   Expected: {test_images_source}")
        print(f"   Please check your dataset folder '{data_yaml_dir}' to confirm.")
        print("   Could not run predictions. Aborting this step.")
        return

    print(f"‚úÖ Found test images source at: {test_images_source}")
    # --- [END OF FIX] ---

    # Define a new directory to save all individual predictions
    pred_save_dir = os.path.join(test_dir, "all_test_predictions")
    os.makedirs(pred_save_dir, exist_ok=True)

    print(f"Running predictions on all images in: {test_images_source}...")

    # Run predict, using the explicit 'source' argument
    model.predict(
        source=test_images_source,  # <--- Using the corrected, explicit path
        project=pred_save_dir,
        name="predict_run",
        exist_ok=True,
        save=True,  # This tells it to save the annotated images
        save_txt=True, # Also saves bounding box .txt files
        conf=0.25   # You can adjust this confidence threshold
    )

    final_save_path = os.path.join(pred_save_dir, "predict_run")
    print(f"‚úî All predicted images saved to: {final_save_path}")

    # Log a few examples to W&B
    try:
        example_preds = glob.glob(os.path.join(final_save_path, "*.jpg"))
        if example_preds:
            wandb.log({"5. Example Predictions": [wandb.Image(img) for img in example_preds[:10]]})
    except Exception as e:
        print(f"‚ö†Ô∏è W&B logging for example predictions failed: {e}")

    return final_save_path

# ======================================================================
# MAIN
# ======================================================================
# 1. Find the most recent run directory (or train a new one)
run_dir = auto_resume()

# 2. [FIX] Get the run name and initialize W&B *now*.
#    This ensures W&B is active even when resuming a completed run.
run_name = os.path.basename(run_dir)
wandb.init(project=WANDB_PROJECT, name=run_name, resume="allow", config=CONFIG)

# 3. Run the visualization & logging
test_dir = run_test_and_visualize(run_dir)

# 4. Run the prediction saving (This function is now defined)
save_all_test_predictions(run_dir, test_dir)

# 5. Finish
print("\nüéâ FINAL COPRA PIPELINE COMPLETE")
print(f"üìÇ All results saved in: {test_dir}")
wandb.finish()
```
